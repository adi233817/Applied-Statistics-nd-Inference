1.  What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

-   In hypothesis testing, Type I and Type II errors are the two possible mistakes that can occur when making decisions about a population based on sample data. Let’s understand them clearly

1. Type I Error (False Positive)

-  Definition: Rejecting a true null hypothesis (H₀).

-  Meaning: You conclude that an effect or difference exists when in reality it does not.

-  Symbol: α (alpha)

Example:
Suppose a court trial —

-  Null hypothesis (H₀): The defendant is innocent.

-  Type I error: Concluding the defendant is guilty when they are actually innocent.

-  Impact: Leads to wrongly detecting an effect; can result in unnecessary actions or costs.

2. Type II Error (False Negative)

-  Definition: Failing to reject a false null hypothesis.

-  Meaning: You conclude there is no effect when in reality there is one.

-  Symbol: β (beta)

Example: In the same court example —

-  Null hypothesis (H₀): The defendant is innocent.

-  Type II error: Concluding the defendant is innocent when they are actually guilty.

-  Impact: Leads to missing a real effect; can result in loss of opportunity or risk.


2.    What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

-  The p-value in hypothesis testing is the probability of getting results as extreme as, or more extreme than, the results actually observed, assuming that the null hypothesis (H₀) is true. In simple words, it tells us how likely our sample data would occur if there were really no effect or no difference in the population.

When interpreting the p-value, a small p-value (typically less than 0.05) suggests that the observed data are unlikely under the null hypothesis. This gives us strong evidence against H₀, so we reject the null hypothesis and accept that there may be a significant effect or difference.

On the other hand, a large p-value (greater than 0.05) means that the data are consistent with the null hypothesis, so we fail to reject H₀. This doesn’t prove that H₀ is true, but it indicates that there isn’t enough evidence to say it’s false.

In short, the p-value helps measure the strength of evidence against the null hypothesis:


-  A low p-value means strong evidence against H₀.


-  A high p-value means weak evidence against H₀.

3. Explain the difference between a Z-test and a T-test, including when to use
each.  

-  The Z-test and T-test are both statistical tests used to compare sample data and make inferences about a population, but they differ mainly in the sample size and whether the population standard deviation is known.

Z-test

A Z-test is used when:

-  The sample size is large (usually n > 30).

-  The population standard deviation (σ) is known.

It relies on the standard normal distribution (Z-distribution).
This test is generally used for comparing sample and population means or proportions when data are normally distributed.

Example:
If you know the population standard deviation and want to test whether the average height of students in a college differs from the national average, you can use a Z-test.

T-test

A T-test is used when- :

-  The sample size is small (usually n ≤ 30).

-  The population standard deviation (σ) is unknown, and you use the sample standard deviation (s) instead.

It relies on the Student’s t-distribution, which is similar to the normal distribution but has thicker tails to account for extra uncertainty in small samples.

Example:
If you don’t know the population standard deviation and want to check whether the average marks of 20 students differ from a given standard value, you would use a T-test.

4.  What is a confidence interval, and how does the margin of error influence
its width and interpretation?

-  A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.

For example, if you say the average height of students is 170 ± 3 cm at a 95% confidence level, it means you are 95% confident that the true average height of all students lies between 167 cm and 173 cm.

-  Margin of Error and Its Influence

The margin of error (ME) is the amount added and subtracted from the sample estimate to create the confidence interval. It depends on:

1.  Sample size (n) – Larger samples reduce the margin of error.

2.  Variability in data (standard deviation) – More variability increases the margin of error.

3.  Confidence level – Higher confidence (e.g., 99% vs 95%) means a wider interval because we want to be more certain.

Effect on Width and Interpretation

-  Larger margin of error → Wider confidence interval.
This means less precision but more confidence that the true value lies within the range.

-  Smaller margin of error → Narrower confidence interval.
This means more precision but less confidence.

5.  Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

-  The purpose of an ANOVA test (Analysis of Variance) is to find out whether there are any significant differences between the means of three or more independent groups. In simple words, it helps us check if all the groups have similar averages or if at least one group’s average is different. For example, if you want to compare the performance of students taught with three different teaching methods, ANOVA can tell you if the teaching method has an effect on their average scores.


ANOVA is an extension of the t-test, which is used to compare only two groups. Instead of doing many t-tests (which increases the chance of errors), ANOVA allows us to test all the groups at once in a single analysis.



For ANOVA to give correct results, certain assumptions must be followed.
First, the samples should be independent, meaning the data collected from one group should not influence another group.
Second, the data in each group should follow a normal distribution.
Third, all groups should have equal variances, which means the spread of data in each group should be roughly the same.


In hypothesis testing, ANOVA sets up two statements. The null hypothesis says that all group means are equal, while the alternative hypothesis says that at least one group mean is different. ANOVA calculates an F-value, which compares the variation between groups to the variation within groups. If this F-value is high and the p-value is less than 0.05, we reject the null hypothesis, meaning that not all group means are the same.


In short, ANOVA helps researchers and analysts test differences among more than two groups in one go, while keeping the results statistically reliable.

6.  Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

- A one-sample Z-test is used to check whether the sample mean is significantly different from a known population mean when the population standard deviation is known.

For example:
Suppose the average height of students in a school is 165 cm. You take a sample of students and want to check if their average height is significantly different from 165 cm.

```python
import numpy as np
from scipy import stats

# Given data
data = [167, 170, 172, 169, 165, 168, 171, 173, 170, 166]  # sample data
population_mean = 165  # known population mean
population_std = 4     # known population standard deviation

# Calculate sample mean
sample_mean = np.mean(data)
n = len(data)

# Calculate Z statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Display results
print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis – sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis – no significant difference between sample and population mean.")
```

Explanation:

1.  We import the required libraries: numpy for math and scipy.stats for statistical functions.

2.  We take a small dataset of sample values.

3.  We know the population mean and population standard deviation.

4.  We calculate the Z-score, which shows how many standard deviations the sample mean is away from the population mean.

5. We find the p-value to see the probability of observing such a result if the null hypothesis is true.

6. If the p-value is less than 0.05, we reject the null hypothesis, meaning the sample mean is significantly different.

8.  Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

-  The Central Limit Theorem (CLT) states that even if the population distribution is not normal, the sampling distribution of the sample means will become approximately normal when the sample size is large enough.

If you take many random samples from a non-normal population and calculate their means, those means will follow a bell-shaped normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a non-normal distribution (Exponential)
population = np.random.exponential(scale=2.0, size=100000)

# Step 2: Draw multiple samples and compute their means
sample_means = []
sample_size = 50  # number of observations in each sample
num_samples = 1000  # number of samples

for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Step 3: Plot the population distribution (non-normal)
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(population, bins=40, color='orange', edgecolor='black')
plt.title("Original Population (Exponential Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 4: Plot the sampling distribution of means (approx. normal)
plt.subplot(1,2,2)
plt.hist(sample_means, bins=40, color='skyblue', edgecolor='black')
plt.title("Sampling Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Explanation:

1-  We first generate data from an exponential distribution, which is not normal (it’s skewed to the right).

2-  We then take 1000 samples, each of size 50, and calculate the mean of each sample.

3-  The left histogram shows the original (non-normal) data.

4-  The right histogram shows the distribution of sample means, which looks approximately normal due to the Central Limit Theorem.

9.  Write a Python function to calculate and visualize the confidence interval
for a sample mean.

-  A confidence interval (CI) gives a range of values within which the true population mean is likely to fall.
For example, a 95% confidence interval means we are 95% confident that the actual population mean lies within that range.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    # Calculate sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # Compute the confidence interval
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, df=n-1)
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    # Print the result
    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    # Visualization
    plt.figure(figsize=(8,4))
    plt.errorbar(x=1, y=mean, yerr=margin_of_error, fmt='o', color='blue', capsize=8, markersize=8)
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.xlim(0, 2)
    plt.xticks([])
    plt.ylabel("Mean Value")
    plt.grid(True)
    plt.show()

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)  # sample data
confidence_interval(data, confidence=0.95)
```

Explanation:

1-  The function first calculates:

-  The sample mean (average of data).

-  The standard error (how much sample means vary from the population mean).

2-  It then uses the t-distribution to calculate the margin of error for the given confidence level (like 95%).

3-  The lower and upper bounds of the confidence interval are calculated as:

Mean±Margin of Error

4-  The result is printed and shown visually with an error bar plot, where the dot represents the mean and the line shows the confidence range.

10.  Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

-  A Chi-square Goodness-of-Fit test is used to check whether the observed data (actual counts) fit a particular expected distribution.
It helps us see if there’s a significant difference between what we observe and what we expect.

For example, if you roll a die 60 times, you expect each face (1–6) to appear 10 times. The test checks if your observed results match that expectation.

```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies (actual data)
observed = np.array([8, 12, 11, 9, 10, 10])  # results from rolling a die 60 times

# Step 2: Expected frequencies (theoretical distribution)
expected = np.array([10, 10, 10, 10, 10, 10])  # each face expected 10 times

# Step 3: Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis – observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis – observed data fits the expected distribution.")
```

Explanation:

1-  Observed frequencies represent the actual results from your experiment.

2-  Expected frequencies represent what you would expect if everything were perfectly fair or balanced.

3-  The function chisquare() from scipy.stats calculates two values:

-  Chi-square statistic (χ²): measures how far observed values deviate from expected ones.

-  P-value: tells how likely it is to see such a difference by chance.

4-  If the p-value is less than 0.05, it means the observed results differ significantly from the expected ones.
